```python
import csv
from time import time
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import rainbow
#from matplotlib import offsetbox
import matplotlib
from adjustText import adjust_text
from sklearn import (manifold, datasets, decomposition, ensemble,
             discriminant_analysis, random_projection)

matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
def plot_country_clustering(X,y,label,spacing,title=None):
    colormap = matplotlib.cm.summer(np.linspace(0, 1, max(label)-min(label)))
    x_min, x_max = np.min(X, 0), np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)     
    plt.figure()
    for i in range(X.shape[0]):
        plt.plot(X[i, 0], X[i, 1],marker=".",ms=15,
                 c = colormap[label[i]-min(label)-1])
        #if i % spacing == 0:
        #    plt.text(X[i, 0], X[i, 1], y[i],
        #         fontdict={'weight': 'bold', 'size': 9},
        #            rotation=26)
    texts = [plt.text(X[i, 0], X[i, 1], y[i],
                      fontdict={'name':'Times New Roman', 'weight': 'bold', 'size': 12}) 
                                     for i in range(X.shape[0]) if i % spacing == 0]
    adjust_text(texts)
    #if title is not None:
    #    plt.title(title,y=-0.001)
    plt.xticks([]), plt.yticks([])
```

# Clustering CSLI Scores

```python
countryList_CSLI = []
with open('./tmp/CSLI.csv') as csvfile:
    countryDist_CSLI = csv.reader(csvfile, delimiter=',')
    for row in countryDist_CSLI:
        countryList_CSLI.append(row[0])
        
countryDist_CSLI = np.genfromtxt ('./tmp/CSLI.csv', delimiter=",")
hdi = countryDist_CSLI[:,1]
hdi = hdi.astype(np.int64)
X_CSLI = countryDist_CSLI[:,2:]
n_samples, n_features = X_CSLI.shape
```

```python
tsne_CSLI = manifold.TSNE(n_components=2, perplexity=3,
                     init='pca', random_state=0)
X_tsne_CSLI = tsne_CSLI.fit_transform(X_CSLI)
```

```python
plot_country_clustering(X_tsne_CSLI, countryList_CSLI, hdi, 1,
        "Clustering of Regions based on CSLI-s scores(Colored With HDI)")
plt.savefig('CSLI_clustering.png')
plt.show()
```
